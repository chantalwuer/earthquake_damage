## Clean Attributes map

```python
attributes_map.head()
```

```python
# Lowercase all the values

def lower_strings(sentence):
    return sentence.lower()
	
attributes_map['DISTRICT'] = attributes_map['DISTRICT'].apply(lower_strings)
attributes_map['UNIT_TYPE'] = attributes_map['UNIT_TYPE'].apply(lower_strings)
attributes_map['UNIT_NAME'] = attributes_map['UNIT_NAME'].apply(lower_strings)
```

```python
attributes_map.head()
```

```python
# Create a dictionary of district codes and names

municipality_codes_df = pd.DataFrame.from_dict(municipality_codes, orient='index', columns=['UNIT_NAME'])
municipality_codes_df['UNIT_NAME'] = municipality_codes_df['UNIT_NAME'].apply(lower_strings)
municipality_codes_df.reset_index(inplace=True)
municipality_codes_df.rename(columns={'index': 'UNIT_CODE'}, inplace=True)
municipality_codes_df.head()
```

```python
# Merge municipality codes with attributes map on UNIT_NAME
# Drop nas, unneccecary columns, and change order of columns 

shp_unit_code = attributes_map.merge(municipality_codes_df, on='UNIT_NAME', how='left')
shp_unit_code.dropna(inplace=True)
shp_unit_code.drop(columns=['UNIT_TYPE'], inplace=True)
shp_unit_code = shp_unit_code[['ID', 'PROVINCE', 'DISTRICT', 'DISTRICT_C', 'UNIT_CODE', 'UNIT_NAME', 'SHAPE_LENG',
       'SHAPE_AREA', ]]
shp_unit_code['UNIT_CODE'] = shp_unit_code['UNIT_CODE'].astype(int)
```

#### Merge municipality codes with ward_unit_code

```python
ward_unit_code = wards_keys.merge(municipality_codes_df, on='UNIT_NAME', how='left')
```

### Check if unit_name matches with municipality codes

```python
# check if a municipality is in both datasets
troubled_values = []
for f in municipality_codes_df.UNIT_NAME:
    if f not in wards_keys.UNIT_NAME.unique():
        troubled_values.append(f)
```

###  


```python
ward_unit_code['UNIT_CODE'] = ward_unit_code['UNIT_CODE'].astype('Int64')

ward_unit_code.head()
```

#### Add district codes to df

```python
# Inverse district_codes dictionary to map district codes to names
district_codes_inv = {v: k for k, v in district_codes.items()}
```

```python
ward_unit_code['district_code'] = ward_unit_code['DISTRICT'].map(district_codes_inv)
ward_unit_code.dropna(inplace=True)
```

```python
ward_unit_code.sort_values(by=['UNIT_CODE']).head()
```

#### Create ward_id by appending

```python
ward_id = []

for i, y in ward_unit_code.iterrows():
    if y['WARD'] < 10:
        ward_id.append(str(y['UNIT_CODE']) + '0' + str(y['WARD']))
    else: 
        ward_id.append(str(y['UNIT_CODE']) + str(y['WARD']))

ward_unit_code['ward_id'] = ward_id
ward_unit_code['ward_id'] = ward_unit_code['ward_id'].astype('Int64')
```
