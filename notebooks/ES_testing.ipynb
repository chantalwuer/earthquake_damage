```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.catplot(y="type_toilet_pre_eq", hue="damage_grade",
                data=df, kind="count",
            height=5, aspect=1.25,legend=False,
               order=df['type_toilet_pre_eq'].value_counts().index)

plt.xlabel('Number of Buildings')
plt.ylabel("Floor Type")
plt.title('Comparison on building damage based on floor type')

plt.legend(title='Damage Severity ', loc='lower right', 
           labels=['Severe', 'Major', 'Minor'])
```

```python
sns.catplot(y="type_toilet_post_eq", hue="damage_grade",
                data=df, kind="count",
            height=5, aspect=1.25,legend=False,
               order=df['type_toilet_post_eq'].value_counts().index)

plt.xlabel('Number of Buildings')
plt.ylabel("Toilet Type")
plt.title('Comparison on building damage based on toilet type')

plt.legend(title='Damage Severity ', loc='lower right', 
           labels=['Severe', 'Major', 'Minor'])
```

```python
df_happened_las_12_month=df[["damage_grade","count_death_last_12_months","count_injury_loss_last_12_months","count_education_drop_last_12_months","count_pregnancy_treatment_drop_last_12_months","count_vaccination_drop_last_12_months","count_occupation_change_last_12_months"]]
```

```python
heatmap_df = df_happened_las_12_month[["damage_grade", "count_injury_loss_last_12_months"]].value_counts().reset_index()
heatmap_df = heatmap_df.groupby(['count_injury_loss_last_12_months','damage_grade']).sum()
heatmap_df.reset_index(inplace=True)
heatmap_df = heatmap_df.rename(columns={0:'values'})
heatmap_df
```
